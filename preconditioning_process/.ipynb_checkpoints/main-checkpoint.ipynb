```python
# video to image 
import cv2
# videoFileList
videoFiles =['resources/KakaoTalk_20201106_134715974.mp4']
videoFilesNum = len(videoFiles)
videoName = videoFiles[0]
path = 'resources/frames/'
for i in range(videoFilesNum):
    print('for문')
    videoName = videoFiles[i]
    cam = cv2.VideoCapture(videoName)
    currentFrame = 0
    if(cam.isOpened()==False):
        print("false")
    else:
        while(True):
            print('while 문')
            ret, frame = cam.read()
            if ret:
                cv2.imwrite(str(path)+'frame%d.jpg' %currentFrame , frame)
                currentFrame = currentFrame+1
                #if cv2.waitKey(25):
                #    break;
            else:
                break
        cam.release()
```

```python
# image resizing 
import os
import glob
from PIL import Image
files = glob.glob('resources/frames/*.jpg')

for f in files:
    print("for")
    img = Image.open(f)
    img_resize = img.resize((int(img.width/4),int(img.height/4)))
    title,ext = os.path.splitext(f)
    img_resize.save('resize_'+title+ext)
```
